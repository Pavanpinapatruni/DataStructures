Selection Sort

- *Run the loop till the n-2 index(last but one element)*
- *In each iteration find the min element by running another loop from main loop index to n-1 index*
- *Swap the minimum element with starting element in inner loop*

## Time Complexity

- Main loop - n times
- inner loop (n-1) + (n-2) + .... + 1
- so it can be total like  = sum of n natural numbers
- n(n-1)/2
- Time Complexity - O(n^2)
